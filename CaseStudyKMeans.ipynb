```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mtcars=pd.read_csv('cars-1.csv')
mtcars.head()
```

```python
car_features=mtcars[['mpg','disp','hp']]
car_features.head()
```

```python
from sklearn.cluster import KMeans
k=3
kmeans=KMeans(n_clusters=k,random_state=42)
#fitting of data
kmeans.fit(car_features)
#Getting the cluster labels
labels=kmeans.predict(car_features)
#Cetroid values
centroid=kmeans.cluster_centers_
print(centroid)
x,y,z=[],[],[]
for i in range(k):
  x.append(centroid[i][0])
  y.append(centroid[i][1])
  z.append(centroid[i][2])

```

```python
plt.subplot(1,2,1)
plt.scatter(car_features['mpg'], car_features['disp'], c='black',s=7)
plt.scatter(x, y, marker='*', c='r', s=700)
plt.title('K-Means Clustering (k=3)')
plt.xlabel('mpg')
plt.ylabel('disp')
plt.show()
plt.subplot(1,2,2)
plt.scatter(car_features['mpg'], car_features['hp'], c='black',s=7)
plt.scatter(x, z, marker='*', c='r', s=700)
plt.title('K-Means Clustering (k=3)')
plt.xlabel('mpg')
plt.ylabel('hp')
plt.show()
plt.subplot(1,3,3)
plt.scatter(car_features['disp'], car_features['hp'], c='black',s=7)
plt.scatter(y, z, marker='*', c='r', s=700)
plt.title('K-Means Clustering (k=3)')
plt.xlabel('disp')
plt.ylabel('hp')
plt.show()
```
